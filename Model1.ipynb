```python
df.head()
X=df.drop(columns=["HeartDisease","MaxHR"])
y=df["HeartDisease"]
X
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
label=LabelEncoder()
onehot=OneHotEncoder()
X_label=X.copy()
list=["ChestPainType","Sex","ExerciseAngina","ST_Slope","RestingECG"]
for item in list:
    X_label[item]=label.fit_transform(X_label[item])
X_label
```

```python
## 1--M  0--F
## 0--N  1--Y
##
```

```python
X_onehot=X.copy()
X_onehot=pd.get_dummies(df,columns=["ChestPainType","Sex","ExerciseAngina","ST_Slope","RestingECG"],dtype=int)
X_onehot
```

```python
y_onehot=X_onehot["HeartDisease"]
X_onehot=X_onehot.drop(columns=["MaxHR","HeartDisease"])
X_onehot
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_onehot[["Age","RestingBP","Cholesterol","Oldpeak"]]=scaler.fit_transform(X_onehot[["Age","RestingBP","Cholesterol","Oldpeak"]])
X_onehot
```

```python
X_label=scaler.fit_transform(X_label)
X_label
```

```python
from sklearn.model_selection import train_test_split
X_train_label, X_test_label, y_train_label, y_test_label = train_test_split(
    X_label, y, test_size=0.25, random_state=42)
X_train_onehot, X_test_onehot, y_train_onehot, y_test_onehot = train_test_split(
    X_onehot, y, test_size=0.25, random_state=42)
X_train_onehot
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,precision_score
```

```python
model_lr_onehot=LogisticRegression()
model_lr_onehot.fit(X_train_onehot,y_train_onehot)
y_pred_onehot=model_lr_onehot.predict(X_test_onehot)
acc1=accuracy_score(y_test_onehot,y_pred_onehot)

model_lr_label=LogisticRegression()
model_lr_label.fit(X_train_label,y_train_label)
y_pred_label=model_lr_label.predict(X_test_label)
acc2=accuracy_score(y_test_label,y_pred_label)
print(f"label {acc1} \n onehot {acc2}")
print(classification_report(y_test_onehot, y_pred_onehot))
```

```python
models=[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),DecisionTreeClassifier(),SVC()]
for model in models:
    curr_model=model
    curr_model.fit(X_train_label,y_train_label)
    y_pred_label=curr_model.predict(X_test_label)
    acc=accuracy_score(y_test_label,y_pred_label)
    pre=precision_score(y_test_label,y_pred_label)
    print(f"{model} :\nAccuracy - {acc}   Precision - {pre}")
```

```python
for model in models:
    curr_model=model
    curr_model.fit(X_train_onehot,y_train_onehot)
    y_pred_onehot=curr_model.predict(X_test_onehot)
    acc2=accuracy_score(y_test_onehot,y_pred_onehot)
    pre2=precision_score(y_test_onehot,y_pred_onehot)
    print(f"{model} :\nAccuracy - {acc2}   Precision - {pre2}")
```

```python
from sklearn.model_selection import cross_val_score
for model in models:
    curr_model=model
    curr_model.fit(X_train_onehot,y_train_onehot)
    y_pred_onehot=curr_model.predict(X_test_onehot)
    score=cross_val_score(curr_model,X_train_onehot,y_train_onehot,cv=5,scoring="accuracy")
    
    acc2=accuracy_score(y_test_onehot,y_pred_onehot)
    pre2=precision_score(y_test_onehot,y_pred_onehot)
    print(f"{model} :\nAccuracy - {acc2}   Precision - {pre2}  Score - {np.mean(score)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(X_train_onehot,y_train_onehot)
y_pred_rf=rf_model.predict(X_test_onehot)

score=cross_val_score(rf_model,X_train_onehot,y_train_onehot,cv=5,scoring="accuracy")


rf_acc=accuracy_score(y_test_onehot,y_pred_rf)
rf_pre=precision_score(y_test_onehot,y_pred_rf)
print(f"{rf_model} :\nAccuracy - {rf_acc}   Precision - {rf_pre}  Score - {np.mean(score)}")
```

```python
from xgboost import XGBClassifier
xgb_model=XGBClassifier(n_estimators=100,max_depth=6,learning_rate=0.5,random_state=42)
xgb_model.fit(X_train_onehot,y_train_onehot)
y_pred_xgb=xgb_model.predict(X_test_onehot)
acc_xgb=accuracy_score(y_test_onehot,y_pred_xgb)
score_xgb=cross_val_score(xgb_model,X_train_onehot,y_train_onehot,cv=5,scoring="accuracy")
pre_xgb=precision_score(y_test_onehot,y_pred_xgb)
print(f"{xgb_model} :\nAccuracy - {acc_xgb}   Precision - {pre_xgb}  Score - {np.mean(score_xgb)}")
```
